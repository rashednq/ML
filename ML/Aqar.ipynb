```python
import pandas as pd
```

```python
df_aqar = pd.read_csv("https://github.com/AFAskar/aqar-scraper/raw/refs/heads/main/aqar_fm_listings.csv")
# show all columns
pd.set_option('display.max_columns', None)
df_aqar.head()
```

```python
print(len(df_aqar.columns))
print(df_aqar.columns)
```

```python
important_features = [
    'area_sqm',
    'num_bedrooms',
    'num_bathrooms',
    'city',
    'district',
    'category_name'
]

target = 'price'

df_model = df_aqar[important_features + [target]]
df_model.shape
```

```python
df_model.head()
```

```python
df_model_sample = df_model.dropna()
```

```python
df_model_sample.head()
```

```python
df_model_sample.isna().sum()
```

```python
x = df_model_sample.drop('price', axis=1)
y = df_model_sample['price']
```

```python
X_encoded = pd.get_dummies(
    x,
    columns=['city', 'district', 'category_name'],
    drop_first=True
)

X_encoded = X_encoded.astype(int)
X_encoded.shape
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X_encoded, y, test_size=0.15, random_state=42
)

val_ratio_within_trainval = 0.15 / 0.85

X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=val_ratio_within_trainval,
    random_state=42
)

X_train.shape, X_val.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)
y_test_pred = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib

def eval_reg(name, y_true, y_pred):
    print(name)
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R2:", r2_score(y_true, y_pred))
    print("-----")

eval_reg("Train", y_train, y_train_pred)
eval_reg("Validation", y_val, y_val_pred)
eval_reg("Test", y_test, y_test_pred)


joblib.dump(model, 'aqar_price_prediction_model.pkl')
model = joblib.load('aqar_price_prediction_model.pkl')
```

```python
x = df_model_sample.drop('price', axis=1)
y = df_model_sample['price']
```

```python
X_encoded = pd.get_dummies(
    x,
    columns=['city', 'district', 'category_name'],
    drop_first=True
)

X_encoded = X_encoded.astype(float)
```

```python
from sklearn.model_selection import train_test_split

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X_encoded, y, test_size=0.15, random_state=42
)

val_ratio_within_trainval = 0.15 / 0.85

X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=val_ratio_within_trainval,
    random_state=42
)
```

```python
from xgboost import XGBRegressor

model1 = XGBRegressor(
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=5,
    gamma=0.1,
    objective='reg:squarederror',
    random_state=42
)

model1.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)
```

```python
y_train_pred = model1.predict(X_train)
y_val_pred   = model1.predict(X_val)
y_test_pred  = model1.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def eval_reg(name, y_true, y_pred):
    print(name)
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R2:", r2_score(y_true, y_pred))
    print("-----")
```

```python
eval_reg("Train", y_train, y_train_pred)
eval_reg("Validation", y_val, y_val_pred)
eval_reg("Test", y_test, y_test_pred)

joblib.dump(model1, 'aqar_price_prediction_xgboost_model.pkl')
model1 = joblib.load('aqar_price_prediction_xgboost_model.pkl')
```
